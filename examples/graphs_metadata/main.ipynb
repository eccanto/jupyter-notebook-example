## Create Graph

```python
import networkx

graph = networkx.Graph()
graph.add_edge('A', 'B', metadata={'steps': ['step 1', 'step 2']})
graph.add_edge('B', 'D', metadata={'steps': ['step 3']})
graph.add_edge('A', 'C', metadata={'steps': ['step 4']})
graph.add_edge('C', 'D', metadata={'steps': ['step 5', 'step 6', 'step 7']})
```

## Get Metadata

```python
graph['A']['C']['metadata']
```

## Plot Graph

```python
from matplotlib import pyplot

networkx.draw(graph, with_labels=True, font_weight='bold')

pyplot.show()
```

## Find paths

```python
list(networkx.all_simple_paths(graph, 'A', 'D'))
```
