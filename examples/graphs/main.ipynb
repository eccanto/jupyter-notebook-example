## Create Graph

```python
import networkx

graph = networkx.Graph()
graph.add_edge('A', 'B')
graph.add_edge('B', 'D')
graph.add_edge('A', 'C')
graph.add_edge('C', 'F')
graph.add_edge('C', 'D')
graph.add_edge('F', 'D')
```

## Plot Graph

```python
from matplotlib import pyplot

networkx.draw(graph, with_labels=True, font_weight='bold')

pyplot.show()
```

## Find paths

```python
list(networkx.all_simple_paths(graph, 'A', 'D'))
```
